```python
"""
Imports:
"""
import pandas
import sklearn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
"""
Load the data:
"""
training_data = pandas.read_csv("../data/dota2TrainingNamed.csv", sep=",")
test_data = pandas.read_csv("../data/dota2TestNamed.csv", sep=",")
```

```python
"""
Check data contents training set:
"""
training_data.head()
```

```python
"""
Seperate training and test data into a class variable and feature variable:
"""
training_features_x = training_data.iloc[:,4:]
training_class_y = training_data["winning_team"]
test_features_x = test_data.iloc[:,4:]
test_class_y = test_data["winning_team"]
```

```python
"""
Create a linear classification model:
"""
classification_linear = SVC(kernel='linear')
classification_linear.fit(training_features_x, training_class_y)
```

```python
"""
Check the performance for the models:
"""
classification_linear_test = classification_linear.predict(test_features_x)
classification_linear_accuracy = accuracy_score(test_class_y, classification_linear_test)
print(classification_linear_accuracy)
```
